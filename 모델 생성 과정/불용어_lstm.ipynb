```python
for i in range(len(df_p2)):
    p = re.sub('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]', '', df_p2['lyric'][i])
    p = re.sub(' +', ' ', p).split()
    l = []
    l2 = ''
    for j in range(len(p)):
        if j != 0 and j % 4 == 0:
            l.append(l2)
            l2 = p[j]
        elif j == 0:
            l2 = l2 + p[j]
        else:
            l2 = l2 + ' ' + p[j]
    l.append(l2)
    df_p2.iloc[i,0] = l
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
stopwords = ['가', '예', '이', '리가', '하', '오', '지', '도', 
             '마치', '와', '우', '거야', '그', '끼', '번','으로','로'
             '에서', '걸','해','고','얄라리','줘요내']
```

```python
x_data = []
for i in range(len(df2)):
    x = df2.iloc[i,2]
    
    for line in x:
        xx = []
        for w in line.split():
            if w not in stopwords:
                xx.append(w)
        x_data.append(' '.join(xx))
x_data[0]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x_data)
```

```python
len(tokenizer.word_index)
```

```python
vocab_size = len(tokenizer.word_index) + 1 # 0 포함 위해
```

```python
sequences = []
for line in x_data:
    encoded = tokenizer.texts_to_sequences([line])[0]
    for i in range(2, len(encoded)): # 단어 둘 이하인 문장은 제외(문장을 만들기 어려우므로) 
        sequence = encoded[:i+1]
        sequences.append(sequence)
len(sequences)
# 43만여개의 문장 샘플이 나온다
# ex) 언제까지 눈치만, 언제까지 눈치만 볼, 언제까지 눈치만 볼 거니 ...
```

```python
tokenizer.word_index['느낌이'], tokenizer.word_index['오잖아']
```

```python
np.unique([len(l) for l in sequences], return_counts = True)
```

```python
sequences = pad_sequences(sequences, maxlen=6, padding='pre')
sequences[0]
```

```python
print(sequences[20:50])
```

```python
# 각 샘플 문장의 마지막 단어를 레이블로 분류하기
sequences = np.array(sequences)
X = sequences[:,:-1]
y = sequences[:,-1]
```

```python
X[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.models import load_model
```

```python
del m
```

```python
m = Sequential()
m.add(Embedding(vocab_size, 10))
m.add(LSTM(128, return_sequences=True))
m.add(LSTM(1024))
m.add(Dense(vocab_size, activation = 'softmax'))
m.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics = 'accuracy')
```

```python
ckp = ModelCheckpoint(filepath = '/content/drive/MyDrive/프로젝트/모델/LSTM8p.h5', monitor = 'loss', save_best_only = True)
m.fit(X,y, epochs = 20, callbacks = [ckp])
```
